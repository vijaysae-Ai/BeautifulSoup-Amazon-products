```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
content=requests.get(url)
```

```python
type(content)
```

```python
soup=BeautifulSoup(content.content,'html.parser')
```

```python
soup
```

```python
lin=soup.find_all('div',attrs={'class':"_2kHMtA"})
```

```python
lin
```

```python
link=[]
pro_name=[]
price=[]
website='https://www.flipkart.com'
final=[]
```

```python
for items in lin:
    href=items.find('a')['href']
    link.append(href)
    name=items.find('div',attrs={'class':"_4rR01T"}).text.strip()
    pro_name.append(name)
    prices=items.find('div',attrs={'class':"_25b18c"})
    price.append(prices.text.split('₹'))
    final.append(website+href)
```

```python
dataa={'title':pro_name,'price':price,'link':final}
```

```python
price
```

```python
dataa
```

```python
df=pd.DataFrame(dataa)
```

```python
df.to_csv('flipkart-web.csv')
```
